### Wage - National and County

### own_code: https://www.bls.gov/cew/classifications/ownerships/sic-ownership-titles.htm
```
Code	Ownership Title
0	Total Covered
5	Private
4	International Government
3	Local Government
2	State Government
1	Federal Government
```
### agglevel_code: https://www.bls.gov/cew/classifications/aggregation/agg-level-titles.htm
```
Level Code	Aggregation Level	Notes
13	National, by Supersector -- by ownership sector	(b,d)
43	MSA, Private, by Supersector	Private Sector Only (d,g)
53	Statewide, by Supersector -- by ownership sector	(d)
73	County, by Supersector -- by ownership sector	(d)
```

````python
for year in ['20', '21', '22']: 
    wages_df=pandas.read_csv(f'source/BLS/employment_wages_by_industry/1026_20{year}.csv')
    # get national level
    wages_df=wages_df[wages_df['agglvl_code']==73]
    # drop unneeded columns
    wages_df=wages_df.drop(columns=['agglvl_code', 'size_code', 'qtr', 'disclosure_code'])
    wages_df.to_csv(f'kl_data/wage_data/county_wage_1026_20{year}.csv')
    display(wages_df.head())
````

````python
# get national wage information per year
for year in ['20', '21', '22']: 
    industry_wage_df=[]
    for each_industry in industry_list: 
        wages_df=pandas.read_csv(f'source/BLS/employment_wages_by_industry/{each_industry}_20{year}.csv')
        # get county level
        wages_df=wages_df[wages_df['agglvl_code']==13]
        # drop unneeded columns
        wages_df=wages_df.drop(columns=['agglvl_code', 'size_code', 'qtr', 'disclosure_code'])
        industry_wage_df.append(wages_df)
    national_wage_df=pandas.concat(industry_wage_df)
    national_wage_df.to_csv(f'kl_data/wage_data/national_wage_by_industry_20{year}.csv')
    display(national_wage_df.head())
````

### Employment

````python
for file_name in ['20', '21', '22']: 
    # drop extra leading rows
    employment_df=pandas.read_excel(f'source/BLS/employment/laucnty{file_name}.xlsx', skiprows=5)
    # drop null column
    employment_df=employment_df.drop(columns=['Unnamed: 5'])
    # drop extra trailing rows
    employment_df=employment_df.iloc[:-3]
    # rename columns
    employment_df.columns=['LAUS_code', 'state_fips', 'county_fips', 'name', 'year', 'labor_force', 'employed', 'unemployed', 'unemployment_rate']
    # make fips columns
    employment_df['fips']=employment_df['LAUS_code'].str[2:7]
    # remove extra columns
    employment_df=employment_df.drop(columns=['LAUS_code', 'state_fips', 'county_fips'])
    employment_df.to_csv(f'kl_data/employment_data/employment_20{file_name}.csv', index=False)
    display(employment_df.head())
````
